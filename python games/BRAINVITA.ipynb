```python
import pygame
import numpy as np
import sys
pygame.init()
wind = pygame.display.set_mode((700,700))
array = np.zeros((7,7))
array[2][0] = 2
array[3][0] = 2
array[4][0] = 2
array[2][1] = 2
array[3][1] = 2
array[4][1] = 2
array[2][2] = 2
array[3][2] = 2
array[4][2] = 2
array[2][3] = 2
array[3][3] = 1
array[4][3] = 2
array[2][4] = 2
array[3][4] = 2
array[4][4] = 2
array[2][5] = 2
array[3][5] = 2
array[4][5] = 2
array[2][6] = 2
array[3][6] = 2
array[4][6] = 2
array[0][3] = 2
array[0][2] = 2
array[0][4] = 2
array[1][3] = 2
array[1][2] = 2
array[1][4] = 2
array[5][3] = 2
array[5][2] = 2
array[5][4] = 2
array[6][3] = 2
array[6][2] = 2
array[6][4] = 2
def draw_board():
    pygame.draw.circle(wind,(175,0,0),(350,350),400)
    for i in range(7):
        for j in range(7):
            if array[i][j] == 1:
                pygame.draw.circle(wind,(100,0,0),(i*100 + 50,j*100 + 50),30)
                pygame.draw.circle(wind,(0,0,0),(i*100 + 50,j*100 + 50),15)
            if array[i][j] == 2:
                pygame.draw.circle(wind,(0,100,100),(i*100 + 50,j*100 + 50),30)
def move(xpos,ypos,xpos1,ypos1):
    if array[xpos1][ypos1] == 1:
        if xpos == xpos1:
            if ypos1 - ypos == 2 and array[xpos][ypos + 1] == 2:
                array[xpos][ypos] = 1
                array[xpos][ypos+1] = 1
                array[xpos][ypos1] = 2
            if ypos - ypos1 == 2 and array[xpos][ypos1 + 1] == 2:
                array[xpos][ypos] = 1
                array[xpos][ypos1+1] = 1
                array[xpos][ypos1] = 2
        if ypos == ypos1:
            if xpos1 - xpos == 2 and array[xpos+1][ypos] == 2:
                array[xpos][ypos] = 1
                array[xpos+1][ypos] = 1
                array[xpos1][ypos1] = 2
            if xpos - xpos1 == 2 and array[xpos1+1][ypos1] == 2:
                array[xpos][ypos] = 1
                array[xpos1+1][ypos1] = 1
                array[xpos1][ypos1] = 2
        draw_board()
    pygame.display.update()
    moves = check_win()
    return moves
def check_win():
    moves = "illegal"
    for i in range(6):
        for j in range(6):
            if array[i][j] == 2:
                if array[i+1][j] == 2 or array[i][j+1] == 2:
                    moves = "legal"
    return moves
                
draw_board()
gameover = False
while gameover == False:
    pygame.display.update()
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            sys.exit()
        if events.type == pygame.MOUSEBUTTONDOWN:
            xpos = int(events.pos[0]/100)
            ypos = int(events.pos[1]/100)
        if events.type == pygame.MOUSEBUTTONUP:
            xpos1 = int(events.pos[0]/100)
            ypos1 = int(events.pos[1]/100)
            moves = move(xpos,ypos,xpos1,ypos1)
            no_of_pieces = 0
            for i in range(7):
                for j in range(7):
                    if array[i][j] == 2:
                        no_of_pieces += 1
            if moves == "legal":
                pygame.display.set_caption("{} PIECES LEFT".format(no_of_pieces))
            if moves == "illegal":
                if no_of_pieces == 1:
                    pygame.display.set_caption("{} PIECE LEFT.YOU ARE A GENIUS".format(no_of_pieces))
                if no_of_pieces == 2 or no_of_pieces == 3:
                    pygame.display.set_caption("{} PIECES LEFT.YOU ARE INTELLIGENT".format(no_of_pieces))
                if no_of_pieces >3:
                    pygame.display.set_caption("{} PIECES LEFT.YOU SHOULD PRACTISE".format(no_of_pieces))
                sys.exit()
```
