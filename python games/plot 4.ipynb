```python
import numpy as np
import pygame
import sys
import math
rowno = 6
colno = 7
blue = (0,0,255)
black = (0,0,0)
red = (255,0,0)
yellow = (255,255,0)


sqsize = 100
radius = int(sqsize/2-5)
width = colno*100
height = (rowno+1) * sqsize
size = (width,height)
def createboard():
    board = np.zeros((rowno,colno))
    return board
def droppiece(board,row,col,piece):
    board[row][col] = piece
def isvalid(board,col):
    return board[rowno-1][col] == 0
def nextrow(board,col):
    for r in range(rowno):
        if board[r][col] == 0:
            return r
def printboard(board):
    print(np.flip(board,0))
def wingame(board,piece):
    for c in range(colno-3):
        for r in range(rowno):
            if board[r][c] == piece and board[r][c+1] == piece and board[r][c+2] == piece and board[r][c+3] == piece:
                return True
    for c in range(colno):
        for r in range(rowno-3):
            if board[r][c] == piece and board[r+1][c] == piece and board[r+2][c] == piece and board[r+3][c] == piece:
                return True
    for c in range(colno-3):
        for r in range(rowno-3):
            if board[r][c] == piece and board[r+1][c+1] == piece and board[r+2][c+2] == piece and board[r+3][c+3] == piece:
                return True
    for c in range(colno-3):
        for r in range(3,rowno):
            if board[r][c] == piece and board[r-1][c+1] == piece and board[r-2][c+2] == piece and board[r-3][c+3] == piece:
                return True
def getboard(board):
    for c in range(colno):
        for r in range(rowno):
            pygame.draw.rect(wind,blue,(sqsize*c,sqsize*r+sqsize,sqsize,sqsize))
            pygame.draw.circle(wind,black,(int(sqsize*c+sqsize/2),int(sqsize*r+sqsize+sqsize/2)),radius)
    for c in range(colno):
        for r in range(rowno):
            if board[r][c] == 1:
                pygame.draw.circle(wind,red,(int(sqsize*c+sqsize/2),height-int(sqsize*r+sqsize/2)),radius)
            elif board[r][c] == 2:
                pygame.draw.circle(wind,yellow,(int(sqsize*c+sqsize/2),height-int(sqsize*r+sqsize/2)),radius)
    pygame.display.update()
board = createboard()
printboard(board)
gameover = False
turn = 0
pygame.init()
wind = pygame.display.set_mode(size)
getboard(board)
pygame.display.update()
myfont = pygame.font.SysFont("monospace",75)
while not gameover:
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            sys.exit()
        if events.type == pygame.MOUSEMOTION:
            pygame.draw.rect(wind,black,(0,0,width,sqsize))
            xpos = events.pos[0]
            if turn == 0:
                pygame.draw.circle(wind,red,(xpos,int(sqsize/2)),radius)
            else:
                pygame.draw.circle(wind,yellow,(xpos,int(sqsize/2)),radius)
        pygame.display.update()
        if events.type == pygame.MOUSEBUTTONDOWN:
            if turn == 0:
                xpos = events.pos[0]
                col = int(math.floor(xpos/sqsize))
                if isvalid(board,col):
                    row = nextrow(board,col)
                    droppiece(board,row,col,1)
                    if wingame(board,1):
                        label = myfont.render("ChickenDinner-1",1, red)
                        wind.blit(label,(40,10))
                        gameover = False
                        getboard(board)
                        pygame.time.wait(3000)
                        sys.exit()
                    else:
                        turn += 1
                        turn = turn %2
            else:
                xpos = events.pos[0]
                col = int(math.floor(xpos/sqsize))
                if isvalid(board,col):
                    row = nextrow(board,col)
                    droppiece(board,row,col,2)
                    if wingame(board,2):
                        label = myfont.render("ChickenDinner-2",1, yellow)
                        wind.blit(label,(40,10))
                        gameover = False
                        getboard(board)
                        pygame.time.wait(3000)
                        sys.exit()
                    else:
                        turn += 1
                        turn = turn %2
            getboard(board)
```
