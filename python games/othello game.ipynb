```python
import pygame
import numpy as np
import math
import sys
pygame.init()
wind = pygame.display.set_mode((400,400))
pygame.Surface.fill(wind,(0,0,0))
array = np.zeros((8,8))
array[3][4] = 2
array[4][3] = 2
array[3][3] = 1
array[4][4] = 1
r = 25
turn = 2
none = True
def get_board():
    for i in range(8):
        for j in range(8):
            pygame.draw.rect(wind,(0,150,0),(i*400/8,j*400/8,400/8-2,400/8-2))
def draw_board():
    for i in range(8):
        for j in range(8):
            if array[j][i] == 1:
                pygame.draw.circle(wind,(255,255,255),(i*50 + r,j*50 + r),22)
            if array[j][i] == 2:
                pygame.draw.circle(wind,(0,0,0),(i*50 + r,j*50 + r),22)
def move(xpos,ypos,update):
    global array
    global turn
    turns = "illegal"
    if array[ypos][xpos] == 0:
        for i in range(xpos + 1,8):
            if array[ypos][i] == 0:
                break
            if array[ypos][i] == turn:
                if (i-xpos)**2 == 1:
                    break
                else:
                    if update == "yes":
                        for j in range(xpos,i):
                            array[ypos][j] = turn
                    turns = "legal"
                    break
        for i in reversed(range(xpos)):
            if array[ypos][i] == 0:
                break
            if array[ypos][i] == turn:
                if (xpos - i)**2 == 1:
                    break
                else:
                    if update == "yes":
                        for j in range(i,xpos+1):
                            array[ypos][j] = turn
                    turns = "legal"
                    break
        for i in range(ypos + 1,8):
            if array[i][xpos] == 0:
                break
            if array[i][xpos] == turn:
                if (i-ypos)**2 == 1:
                    break
                else:
                    turns = "legal"
                    if update == "yes":
                        for j in range(ypos,i):
                            array[j][xpos] = turn
                    break
        for i in reversed(range(ypos)):
            if array[i][xpos] == 0:
                break
            if array[i][xpos] == turn:
                if (ypos - i)**2 == 1:
                    break
                else:
                    turns = "legal"
                    if update == "yes":
                        for j in range(i,ypos+1):
                            array[j][xpos] = turn
                    break
        for i in range(xpos+1,8):
            if ypos+i-xpos < 8:
                if array[ypos+i-xpos][i] == 0:
                    break
                if array[ypos + i-xpos][i] == turn:
                    if (xpos-i) ** 2 == 1:
                        break
                    else:
                        if update == "yes":
                            for j in range(xpos,i):
                                array[ypos+j - xpos][j] = turn
                        turns = "legal"
                        break
        for i in range(xpos + 1,8):
            if ypos - i + xpos >= 0:
                if array[ypos - i + xpos][i] == 0:
                    break
                if array[ypos - i + xpos][i] == turn:
                    if (xpos-i) ** 2 == 1:
                        break
                    else:
                        if update == "yes":
                            for j in range(xpos,i):
                                array[ypos-j+xpos][j] = turn
                        turns = "legal"
                        break
        for i in reversed(range(xpos)):
            if ypos - i + xpos < 8:
                if array[ypos - i + xpos][i] == 0:
                    break
                if array[ypos - i + xpos][i] == turn:
                    if (xpos - i) ** 2 == 1:
                        break
                    else:
                        if update == "yes":
                            for j in range(i,xpos+1):
                                array[ypos + xpos - j][j] = turn
                        turns = "legal"
                        break
        for i in reversed(range(xpos)):
            if ypos + i - xpos >= 0:
                if array[ypos + i - xpos][i] == 0:
                    break
                if array[ypos + i - xpos][i] == turn:
                    if (xpos-i) ** 2 == 1:
                        break
                    else:
                        if update == "yes":
                            for j in range(i,xpos + 1):
                                array[ypos + j - xpos][j] = turn
                        turns = "legal"
                        break
    if turns == "legal" and update == "yes":
        if turn == 2:
            turn = 1
        else:
            turn = 2
    return turns

def count():
    no_of_1 = 0
    no_of_2 = 0
    for i in range(8):
        for j in range(8):
            if array[i][j] == 1:
                no_of_1 += 1
            if array[i][j] == 2:
                no_of_2 += 1
    if no_of_1>no_of_2:
        w = 1
    else:
        w = 2
    return w
            
pygame.display.set_caption("PLAYER 2 TURN")                                                                                                                                         
gameover = False
get_board()
draw_board()
while gameover == False:
    
    pygame.display.update()
    
        
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            sys.exit()
            pygame.quit()
        if events.type == pygame.MOUSEBUTTONDOWN:
            pygame.display.set_caption("PLAYER {} TURN".format(turn))
#             none = True
#             for i in range(8):
#                 for j in range(8):
#                     if array[i][j] == 0:
#                         none = False
#             if none == True:
#                 winner = count()
#                 pygame.display.set_caption("PLAYER {} WINS".format(winner))
#                 gameover = True
            xpos = int(events.pos[0]/50)
            ypos = int(events.pos[1]/50)
            move(xpos,ypos,"yes")
            flag="illegal"
            for i1 in range(8):
                for j1 in range(8):
                    if array[j1][i1] == 0:
                        flag1 = move(i1,j1,"no")
                        if flag1 == "legal":
                            flag=flag1
                            break
                if flag=="legal":
                    break
            if flag=="illegal":
                if turn == 2:
                    turn = 1
                else:
                    turn = 2
                for i1 in range(8):
                    for j1 in range(8):
                        if array[j1][i1] == 0:
                            flag1 = move(i1,j1,"no")
                            if flag1 == "legal":
                                flag=flag1
                                break
                    if flag=="legal":
                        break
                if flag == "illegal":
                    winner = count()
                    pygame.display.set_caption("PLAYER {} WINS".format(winner))
                    gameover = True
                        
#             flag1 = check(turn)
#             if flag1 == "illegal":
#                 if turn == 2:
#                     turn = 1
#                 else:
#                     turn = 2
#                 flag2 = check(turn)
#                 if flag2 == "illegal":
#                     winner = count()
#                     pygame.display.set_caption("PLAYER {} WINS".format(winner))
#                     gameover = True
                    
        draw_board()
            
        
```
