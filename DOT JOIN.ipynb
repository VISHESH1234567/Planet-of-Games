```python
import pygame
import numpy as np
import sys
import math
pygame.init()
moves = []
dots = []
wind = pygame.display.set_mode((620,620))
pygame.Surface.fill(wind,(255,255,255))
array = np.zeros((7,7))
def get_board():
    global turn
    for i in range(7):
        for j in range(7):
            pygame.draw.circle(wind,(0,0,0),(j*100+10,i*100+10),10)
            
    pygame.display.update()
def move(xpos,ypos,xpos1,ypos1):
    global turn
    dot1 = ypos*7 + xpos
    dot2 = ypos1*7 + xpos1
    if max(dot1,dot2) - min(dot1,dot2) == 1 and (min(dot1,dot2),max(dot1,dot2)) not in moves:
        moves.append((min(dot1,dot2),max(dot1,dot2)))
        pygame.draw.rect(wind,(0,0,0),(min(xpos,xpos1)*100+12,ypos*100+5, 90,10))
        turn+= 1
    if max(dot1,dot2) - min(dot1,dot2) == 7 and (min(dot1,dot2),max(dot1,dot2)) not in moves:
        moves.append((min(dot1,dot2),max(dot1,dot2)))
        pygame.draw.rect(wind,(0,0,0),(xpos*100+5,min(ypos,ypos1)*100+12, 10,90))
        turn += 1
    return turn
    return moves
turn = 1
red_sq = 0
blue_sq = 0
def draw_sq():
    sq_made = False
    global turn
    global red_sq
    global blue_sq
    for i in range(6):
        for j in range(6):
            dot = i*7 +j
            if (dot,dot+1) in moves and (dot,dot+7) in moves and (dot+1,dot+8) in moves and (dot+7,dot+8) in moves and dot not in dots:
                if turn % 2 == 0:
                    pygame.draw.rect(wind,(255,0,0),(j*100+15,i*100+15, 90,90))
                    red_sq += 1
                    sq_made = True
                else:
                    pygame.draw.rect(wind,(0,0,255),(j*100+15,i*100+15, 90,90))
                    blue_sq += 1
                    sq_made = True
                dots.append(dot)
    if sq_made == True:
        turn += 1
                
    if len(moves) == 84:
        if red_sq > blue_sq:
            winner = 1
        elif red_sq < blue_sq:
            winner = 2
        else:
            pygame.display.set_caption("MATCH IS TIE")
            pygame.time.wait(2000)
            sys.exit()
        pygame.display.set_caption("PLAYER {} WINS".format(str(winner)))
        pygame.time.wait(2000)
        sys.exit()
        
    return moves

gameover = False
get_board()
take = True
while gameover == False:
    if turn%2 == 1:
        pygame.display.set_caption("PLAYER {} TURN".format(str(turn%2)))
    else:
        pygame.display.set_caption("PLAYER {} TURN".format(str(turn%2+2)))
    pygame.display.update()
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            sys.exit()
        if events.type == pygame.MOUSEBUTTONDOWN:
            if take == True:
                xpos = int(events.pos[0]/(620/7))
                ypos = int(events.pos[1]/(620/7))
                if turn % 2 == 1:
                    pygame.draw.circle(wind,(255,0,0),(xpos*100+10,ypos*100+10),10)
                else:
                    pygame.draw.circle(wind,(0,0,255),(xpos*100+10,ypos*100+10),10)
                take = False
            elif take == False:
                xpos1 = int(events.pos[0]/(620/7))
                ypos1 = int(events.pos[1]/(620/7))
                if turn % 2 == 1:
                    pygame.draw.circle(wind,(255,0,0),(xpos1*100+10,ypos1*100+10),10)
                else:
                    pygame.draw.circle(wind,(0,0,255),(xpos1*100+10,ypos1*100+10),10)
                move(xpos,ypos,xpos1,ypos1)
                draw_sq()
                take = True
                pygame.draw.circle(wind,(0,0,0),(xpos*100+10,ypos*100+10),10)
                pygame.draw.circle(wind,(0,0,0),(xpos1*100+10,ypos1*100+10),10)
```
