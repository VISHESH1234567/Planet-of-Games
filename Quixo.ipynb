```python
import pygame
import sys
import numpy as np
pygame.init()
wind = pygame.display.set_mode((750,750))
arr = np.zeros((5,5))

def getboard():
    myfont = pygame.font.SysFont("monospace",150)
    for i in range(5):
        for j in range(5):
            if arr[i][j]==0:
                pygame.draw.rect(wind,(255, 192, 0),(j*150+5,i*150+5,140,140))
            elif arr[i][j]==1:
                pygame.draw.rect(wind,(218, 165, 32),(j*150+5,i*150+5,140,140))
                label = myfont.render("O",1,(92, 64, 51))
                wind.blit(label,(j*150+30,i*150-5))
            else:
                pygame.draw.rect(wind,(218, 165, 32),(j*150+5,i*150+5,140,140))
                label = myfont.render("X",1,(92, 64, 51))
                wind.blit(label,(j*150+30,i*150-5))
    pygame.display.update()
def move_piece(xpos,ypos,direction,turn):
    move = False
    if direction==0:
        for j in range(xpos,4):
            move = True
            arr[ypos][j] = arr[ypos][j+1]
        if move:
            arr[ypos][4] = turn
    elif direction==1:
        for j in range(xpos,0,-1):
            move = True
            arr[ypos][j] = arr[ypos][j-1]
        if move:
            arr[ypos][0] = turn
    elif direction==2:
        for i in range(ypos,4):
            move = True
            arr[i][xpos] = arr[i+1][xpos]
        if move:
            arr[4][xpos] = turn
    else:
        for i in range(ypos,0,-1):
            move = True
            arr[ypos][i] = arr[ypos][i-1]
        if move:
            arr[0][xpos] = turn
    return move

def check_win(turn):
    for i in range(5):
        if arr[i][0]==turn*(-1) and arr[i][1]==turn*(-1) and arr[i][2]==turn*(-1) and arr[i][3]==turn*(-1) and arr[i][4]==turn*(-1):
            return turn*(-1)
        if arr[0][i]==turn*(-1) and arr[1][i]==turn*(-1) and arr[2][i]==turn*(-1) and arr[3][i]==turn*(-1) and arr[4][i]==turn*(-1):
            return turn*(-1)
    if arr[0][0]==turn*(-1) and arr[1][1]==turn*(-1) and arr[2][2]==turn*(-1) and arr[3][3]==turn*(-1) and arr[4][4]==turn*(-1):
        return turn*(-1)
    if arr[0][4]==turn*(-1) and arr[1][3]==turn*(-1) and arr[2][2]==turn*(-1) and arr[1][3]==turn*(-1) and arr[4][0]==turn*(-1):
        return turn*(-1)
    for i in range(5):
        if arr[i][0]==turn and arr[i][1]==turn and arr[i][2]==turn and arr[i][3]==turn and arr[i][4]==turn:
            return turn
        if arr[0][i]==turn and arr[1][i]==turn and arr[2][i]==turn and arr[3][i]==turn and arr[4][i]==turn:
            return turn
    if arr[0][0]==turn and arr[1][1]==turn and arr[2][2]==turn and arr[3][3]==turn and arr[4][4]==turn:
        return turn
    if arr[0][4]==turn and arr[1][3]==turn and arr[2][2]==turn and arr[1][3]==turn and arr[4][0]==turn:
        return turn
    return 0
getboard()
gameover = False
turn = 1
pygame.display.set_caption("PLAYER 1 TURN")
while not gameover:
    pygame.display.update()
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            sys.exit()
        if events.type==pygame.MOUSEBUTTONDOWN:
            selected = False
            xpos = int(events.pos[0]/150)
            ypos = int(events.pos[1]/150)
            if xpos==0 or xpos==4 or ypos==0 or ypos==4:
                if arr[ypos][xpos]==0 or arr[ypos][xpos]==turn:
                    selected = True
        if events.type==pygame.KEYDOWN:
            if selected:
                if events.key == pygame.K_RIGHT:
                    output = move_piece(xpos,ypos,0,turn)
                if events.key == pygame.K_LEFT:
                    output = move_piece(xpos,ypos,1,turn)
                if events.key == pygame.K_UP:
                    output = move_piece(xpos,ypos,3,turn)
                if events.key == pygame.K_DOWN:
                    output = move_piece(xpos,ypos,2,turn)
                if output:
                    getboard()
                    ans = check_win(turn)
                    if ans!=0:
                        pygame.display.set_caption("PLAYER {} WINS".format(int((ans-3)/(-2))))
                        sys.exit()
                    turn*=-1
                    pygame.display.set_caption("PLAYER {} TURN".format(int((turn-3)/(-2))))
                selected = False
```
